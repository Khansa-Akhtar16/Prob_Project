### Basic inspection and cleaning 

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('diabetes_prediction_dataset.csv')
# dataset ki shape wagaira dekhni pehle ke humain pata chale ke dataset me kitne rows aur columns hai, aur total cells kitne hai.
print("shape of the dataset: ", df.shape)
print("total cells of the dataset: ", df.shape[0]*df.shape[1])
```

```python
# ab hum ne columns ke name orr dtypes dekhni han
print("columns name and their data types: \n", df.dtypes)
print('\n')
print(df.info())
```

```python
### now, inspecting the first and last 5 rows of the dataset
print("first 5 rows of the dataset: \n", df.head())
print("\n")
print("last 5 rows of the dataset: \n", df.tail())
```

```python
print("checking for the description of the dataset: \n", df.describe())
```

```python
# ab hum categorical columns ke unique values dekhni han
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"unique values in column '{col}': {df[col].unique()}")
    
```

```python
# now, we've to class balancing stuff
print("value counts of the target variable: \n", df['diabetes'].value_counts())
# as percentage
print("value counts of the target variable in percentage: \n", df['diabetes'].value_counts(normalize=True)*100)
```

```python
# now, we've to count the number of "No Info" 
no_info_count= (df['smoking_history'] == 'No Info').sum()
total = len(df)
print(f"no info count: {no_info_count}")
print(f"Percentage of the dataset: {no_info_count/total *100 :.2f}%")
```

```python
# now, replacing the No Info with unkown, cuz' we don't know whether the person is a smoker or not, so it's better to keep it as unknown rather than removing those rows.
df['smoking_history'] = df['smoking_history'].replace('No Info', 'unknown')
print("unique values in column 'smoking_history' after replacement: ", df['smoking_history'].unique()) 
print("value counts of the 'smoking_history' column after replacement: \n", df['smoking_history'].value_counts())
```

```python
# now, we've to detect the duplicated rows in the dataset
duplicated_rows = df[df.duplicated()]
print("number of duplicated rows: ", duplicated_rows.shape[0])
duplicated = df.duplicated().sum()
print(f"number of duplicated rows: {duplicated/len(df)*100:.2f}%")
print(df[df.duplicated(keep=False)].sort_values(by=list(df.columns)).head(10))
```

```python
df = df.drop_duplicates()
print(f"Rows after removing duplicates: {len(df)}")
```

```python
#now, we're gonna check the age column as it seems suspicious 
print("Suspicious ages (below 1 year):")
print(df[df['age'] < 1]['age'].value_counts())

print(f"\nTotal rows with age < 1: {(df['age'] < 1).sum()}")
```

```python
# now, check the binary columns for any suspicious values
for col in ['hypertension', 'heart_disease', 'diabetes']:
    unique_vals = df[col].unique()
    print(f"{col}: {unique_vals}")
```

```python
#now, check for the BMI boundaries
print("BMI below 10:", (df['bmi'] < 10).sum())
print("BMI above 60:", (df['bmi'] > 60).sum())
print("\nRows with BMI > 60:")
print(df[df['bmi'] > 60][['age', 'bmi', 'diabetes']].head())
```

```python
# full invalid summary
print("Age < 1:", (df['age'] < 1).sum())
print("Age > 120:", (df['age'] > 120).sum())
print("BMI < 10:", (df['bmi'] < 10).sum())
print("BMI > 60:", (df['bmi'] > 60).sum())
print("HbA1c < 3:", (df['HbA1c_level'] < 3).sum())
print("HbA1c > 15:", (df['HbA1c_level'] > 15).sum())
print("Glucose < 50:", (df['blood_glucose_level'] < 50).sum())
print("Glucose > 500:", (df['blood_glucose_level'] > 500).sum())
```

```python
#now, drop the rows where age < 1
df = df[df['age'] >= 1]
print(f"Rows after removing invalid ages: {len(df)}")
```

```python
# now, we do the outlier detection in all the numeric columns and handling
numeric_cols = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)].shape[0]
    print(f"{col}: Q1={Q1}, Q3={Q3}, IQR={IQR:.2f}, Lower={lower:.2f}, Upper={upper:.2f}, Outliers={outliers}")
```

```python
#Cap outliers 
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower, upper)

print("Outlier capping done for numeric columns.")
print(df[numeric_cols].describe())
```
